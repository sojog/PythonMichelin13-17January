```python
import pandas as pd
```

```python
# we read the data from the csv file
# the index is the column "id"
df = pd.read_csv("exemple.csv", index_col="id")
# we display just the first 5 elements
df.head()
```

```python
real = ["Spain", "Real Madrid", 15]
milan = ["Italy", "AC Milan", 7]
bayern = ["Germany", "Bayern Munchen", 6]
```

```python
teams_dict = {
    "country":[ real[0], milan[0], bayern[0] ],
    "team":[ real[1], milan[1], bayern[1] ],
    "trophies":[ real[2], milan[2], bayern[2] ]
}
teams_dict
```

```python
big_guns_df = pd.DataFrame(teams_dict)
big_guns_df
```

```python
list_of_dfs_to_concat = [big_guns_df, df]
pd.concat(list_of_dfs_to_concat).head(5)
```

```python
list_of_dfs_to_concat = [df, big_guns_df]
# ignore_index -> reindexing the newly created dataframe
merged_df = pd.concat(list_of_dfs_to_concat, ignore_index=True)
# The last 5 elements
merged_df.tail()
```

```python
merged_df.sort_values(["trophies", "team", "country"], ascending = [False, True, True])
```

```python
merged_df.sort_values(["trophies", "team", "country"], ascending = [False, True, True], inplace=True)
```

```python
merged_df.reset_index(inplace=True)
```

```python
merged_df = merged_df.drop(axis=1, columns=["index", "level_0"])
merged_df
```

# Starting from index 1 instead of index 0
